```python
v2=np.array([[1,1],[0,2]])
```

v2

```python
v2
```

```python
eig_Val,eig_vect=np.linalg.eig(v2)
```

```python
eig_Val
```

```python
eig_vect
```

```python
c=eig_vect
```

```python
d=np.diag(eig_Val)
```

```python
d
```

```python
c_inv=np.linalg.inv(c)
```

```python
c_inv
```

```python
np.linalg.matrix_power(v2,3)
```

```python
c@np.linalg.matrix_power(d,3)@c_inv
```
