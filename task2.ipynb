```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# --- Load dataset ---
df = pd.read_csv("TitanicDataCleaned.csv")

# Copy dataframe to avoid modifying the original
df_processed = df.copy()

# ====== Imputation ======
# Age -> fill missing with mean
df_processed["Age"] = SimpleImputer(strategy="mean").fit_transform(df_processed[["Age"]])

# Fare -> fill missing with median
df_processed["Fare"] = SimpleImputer(strategy="median").fit_transform(df_processed[["Fare"]])

# Embarked -> fill missing with most frequent
df_processed["Embarked"] = SimpleImputer(strategy="most_frequent").fit_transform(df_processed[["Embarked"]]).ravel()

# Cabin -> fill missing with 'Unknown'
df_processed["Cabin"] = df_processed["Cabin"].fillna("Unknown")

# ====== Scaling ======
scaler = StandardScaler()
num_cols = ["Age", "Fare", "SibSp", "Parch", "Pclass"]
df_processed[num_cols] = scaler.fit_transform(df_processed[num_cols])

# --- Save processed file ---
df_processed.to_csv("TitanicDataProcessed.csv", index=False)

print("✅ Processing Done! File saved as TitanicDataProcessed.csv")
print(df_processed.head())
```
