```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import matplotlib.pyplot as plt
```

```python
from google.colab import files

# لرفع ملف واحد
uploaded = files.upload()
# 1. تحميل البيانات
data = pd.read_csv("Social_Network_Ads.csv")

```

```python
# 2. اختيار الأعمدة المهمة (مثلاً Age و EstimatedSalary للتنبؤ بالشراء)
X = data[["Age", "EstimatedSalary"]]
y = data["Purchased"]
```

```python
# 3. تقسيم البيانات (75% تدريب ، 25% اختبار)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# 4. مقياس القياس (Scaling) لتحسين الأداء
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# 5. إنشاء وتدريب نموذج Logistic Regression
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)
```

```python
# 6. التنبؤ
y_pred = model.predict(X_test)
# 7. التقييم
print("الدقة (Accuracy):", accuracy_score(y_test, y_pred))
print("\nمصفوفة الالتباس (Confusion Matrix):\n", confusion_matrix(y_test, y_pred))
print("\nتقرير التصنيف (Classification Report):\n", classification_report(y_test, y_pred))
```
