```python
import pandas as pd
import os
```

###Merging to one dataset file .csv

```python
path=r"D:\Sales_Analysis\Sales_Data"
files=os.listdir(path)
df=pd.DataFrame()
for file in files :
    full_path=os.path.join(path,file)
    df_temp=pd.read_csv(full_path)
    df=pd.concat([df,df_temp],ignore_index=True)
    
    

```
