```python
df_duplicate_order['products']= df.groupby('Order ID')['Product'].transform(lambda x :','.join(x))
```

```python
df_duplicate_order.drop_duplicates('Order ID',keep='first',inplace=True)
```

```python
df_duplicate_order['products'].str.split(',')
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df_duplicate_order ['products']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```
