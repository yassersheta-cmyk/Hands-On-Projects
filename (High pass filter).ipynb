```python
import cv2
import numpy as np
from skimage import data
import matplotlib.pyplot as plt

# 1.   test  ( )
image = data.camera()  #       scikit-image

# 2.   Gaussian Blur (Low Pass Filter)
low_pass = cv2.GaussianBlur(image, (5, 5), 0)

# 3.   High Pass:  -  
high_pass = image - low_pass

# 4. ()  Sharpening:  +  High Pass
sharpened = image + high_pass

# 5.   
images = [image, low_pass, high_pass, sharpened]
titles = ['original data', 'Low Pass ', 'High Pass ', 'Sharpening']

plt.figure(figsize=(15, 10))
for i in range(len(images)):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
