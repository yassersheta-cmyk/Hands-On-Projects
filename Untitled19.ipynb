```python
import numpy as np
from math import inf
```

```python
v1=np.array([1,2,3,4,5])
v2=np.array([2,12,5,6,4])
```

```python
v1+v2
```

```python
np.linalg.norm(v2,inf)
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
```

```python
y=v1
y_hat=v2
```

```python
error=y-y_hat
```

```python
(np.linalg.norm(error)**2)/v1.shape[0]
```

```python
mse(y,y_hat)
```

```python
v1.shape[0]
```

```python
mae(y,y_hat)
```
